```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
courses = pd.read_csv('Online_Courses.csv')
```

```python
courses.head()
```

```python
courses = courses.drop(columns=['Unnamed: 0'])
```

```python
courses['combined_features'] = courses['Title'] + ' ' + courses['Short Intro'] + ' ' + courses['Skills']
```

```python
courses['combined_features'] = courses['combined_features'].fillna('').str.lower()
```

```python
courses.head()
```

```python
def preprocessed_text (text):
    text = text.lower()
    return text
```

```python
user_text = "Hello how are you here to go"
user_text = preprocessed_text(user_text)
print(user_text)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
course_tfidf = vectorizer.fit_transform(courses['combined_features'])
user_tfidf = vectorizer.transform([user_text])
```

```python
similarity = cosine_similarity(user_tfidf, course_tfidf)
```

```python
top_index = similarity.argsort()[0][-5:][::-1]
recommended_courses = courses.iloc[top_index]
```

```python
print(recommended_courses[['Title', 'URL', 'Short Intro']])
```

```python
import pickle

with open('vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

with open('course_tfidf.pkl', 'wb') as f:
    pickle.dump(course_tfidf, f)
    
courses.to_csv('proprocessed_courses.csv', index = False)
```

```python
with open('vectorizer.pkl', 'rb') as f:
    main_vectorizer = pickle.load(f)
    
with open('course_tfidf.pkl', 'rb') as f:
    course_vector = pickle.load(f)
```

```python
preprocessed_courses = pd.read_csv('proprocessed_courses.csv')
```

```python
def recommend_courses(user_text):
    user_text = user_text.lower()
    user_vector = main_vectorizer.transform([user_text])
    cos_sim = cosine_similarity(user_vector, course_vector)
    top_indices = cos_sim.argsort()[0][-5:][::-1]
    return preprocessed_courses.iloc[top_indices][['Title', 'URL', 'Short Intro']]
```
